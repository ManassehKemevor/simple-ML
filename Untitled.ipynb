```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("breast-cancer.csv")
df.info()
```

```python
df.describe()
```

```python
X=df.drop(["id","diagnosis"],axis=1)
Y=df["diagnosis"]
```

```python
c=LabelEncoder()
y=c.fit_transform(Y)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2,random_state=21)
```

```python
col=X.columns
preprocess=ColumnTransformer(transformers=([
    ("num",StandardScaler(),col)
]
))
```

```python
Xtrain=preprocess.fit_transform(xtrain)
Xtest=preprocess.transform(xtest)
```

```python
model=SVC()
model.fit(Xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
```

```python
accu=accuracy_score(ytest,ypred)
accu
```

```python
parameter={
    'C':[0.1, 1, 10, 100, 1000],
    'gamma': ['scale', 'auto', 0.001, 0.01, 0.1, 1],
    'kernel': ['rbf', 'linear', 'poly']
}
```

```python
mod=GridSearchCV(estimator=SVC(),param_grid=parameter,cv=5,n_jobs=2, verbose=2)
```

```python
mod.fit(Xtrain,ytrain)
```

```python
mod.best_params_
```

```python
bestModel=SVC(C=1,gamma="scale",kernel="rbf")
bestModel.fit(Xtrain,ytrain)
ypred=bestModel.predict(Xtest)
```

```python
accu=accuracy_score(ytest,ypred)
accu
```
